# **Credit Card Fraud Detection**
This project  uses the public dataset available on kaggle to evaluate a model which can predict whether a given transaction is fraud or legit based on given features. The data of credit cards is very private and hence it's features has been converted to numerical values using Principal Component Analysis.


## Importing all the required libraries
We will be importing various **Python** Libraries which are required.

```python
import numpy as np  # We need array to compute all the calculations
import pandas as pd # As the data is not pre processes hence we need pandas to process the data
from sklearn.model_selection import train_test_split # used to split the data
from sklearn.linear_model import LogisticRegression # to use logistics regression
from sklearn.metrics import accuracy_score # for checking accuracy of model
```

```python
# We are now loading the dataset to Pandas Dataframe
credit_card_data = pd.read_csv('/content/drive/MyDrive/Data/creditcard.csv')
```

```python
# Having an overview by using head function to see few rows of whole data
credit_card_data.head()
credit_card_data.tail() # Gives last few rows of dataset
```

The given dataset has time of trasaction in second from first transaction and features from V1 to V28 as the Credit Card data is very sensitve information we cannot use it so the features have been converted into numerical values using principal component analysis method and we havew further amounts in USD and class (label) which means.


*   0 - Real Transaction
*   1 - Fake Transaction







```python
# We then try to get information on dataset using info function
credit_card_data.info()
```

```python
# To check number of missing values in each column by using isnull and sum function
credit_card_data.isnull().sum()
```

```python
# As the dataset has 0 set of amount and 0 class values missing and it is very large dataset we can drop values by using
credit_card_data.dropna(how='any')
```

```python
# Now we are gonna see the number of real and fraud transaction
credit_card_data['Class'].value_counts()
```

### [ Dataset is highly unbalanced ]
Now as the 99 percent of data is real transaction if we give this dataset then the machine learning model won't be able to recognise the fraud transaction and if we give new transaction it will classify as the real transaction. Hence we need to do some more analysis and processing of dataset.

0--> Normal Transaction
1--> Fraud Transaction

```python
# We will seperate the normal and fraud data to better analyze it
normal_data = credit_card_data[credit_card_data.Class == 0] # Getting all normal dataset
fraud_data = credit_card_data[credit_card_data.Class == 1] # Getting all fraud dataset
```

```python
# Checking dimension of both dataset
print(normal_data.shape)
print(fraud_data.shape)
```

```python
# Let us analyze the dataset differently in statistical way
normal_data.Amount.describe()
```

```python
# Fraud dataset
fraud_data.Amount.describe()
```

```python
# We will here try to compute the mean of all features for normal and fraud transaction to see any insights
credit_card_data.groupby('Class').mean()
```

Now we can observe that there is difference in mean of each feature in case of normal and fruad transaction and hence that is useful for our algorithm.

### Under-Sampling
We are gonna take a sample dataset having similar distribution of normal and fruad dataset.

Number of fraud transaction ---> 492

So we will take random sample having 492 values and that will form a good datset to do the analysis

```python
# Sampling random value in normal dataset
normal_data_sample = normal_data.sample(n=492) # randomly sampling 492 data points
```

Now we will combine all the fraud and this new normal datset into one table using Concatenation.

```python
new_dataset = pd.concat([normal_data_sample,fraud_data], axis = 0) # axis refer to whther to add row wise or colums wise so we want row hence axis = 0
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts() # Count number of values based on class
```

This step is important to check whether we have got good dataset after random sampling as it ensures good model.

```python
new_dataset.groupby('Class').mean() # We are grouping dataset on class and finding mean of each feature to compare difference in mean
```

Splitting the dataset into features and targets

X ---> Features
Y ---> Targets

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting whole dataset into Training, Test data to train algorithm



```python
# We are creating 4 arrays having corresponding feature and target and test size indicate percenatage of data in test
# stratify signifies that test and train should have even distribution of 0 and 1 class based on target
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state = 2)
```

```python
# Cheking number of data points
print(X.shape,X_train.shape,X_test.shape)
```

## Train the model

```python
# Using model
model = LogisticRegression()
```

```python
# Training the Logistics Regression Model with Training Dataset
model.fit(X_train,Y_train)
```

## Evaluating the model

Accuracy Score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train) # Predictin the label for training data
training_data_accuracy = accuracy_score(X_train_prediction,Y_train) # Compairing prediction with the actual label
```

```python
# Printing the accuracy on training data
print('The accuracy score on training data is :',training_data_accuracy)
```

```python
# Accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# Accuracy score on train data
print('The accuracy score on test data is :',test_data_accuracy)
```
